```python
import joblib
import pandas as pd
```

**Cargamos modelo y datos a predecir**

```python
model = joblib.load('../modelos/modelo_LGBM_EXO.pkl')
X_prueba = pd.read_csv('../data_sample/X_prueba_EXO.csv')
```

```python
print("LGB best_score:", model.best_score_)
model.best_params_
```

```python
X_prueba
```

**Realizamos la prediccion**

```python
X_prueba_pred = X_prueba.copy()
X_prueba_pred["Prediccion"] = model.predict(X_prueba)
```

**Visualizamos los datos con su prediccion**

```python
pred_mapping = {
    0 : 'FALSE POSITIVE',
    1 : 'CANDIDATE',
    2 : 'CONFIRMED'
}
X_prueba_pred["Prediccion"] = X_prueba_pred["Prediccion"].map(pred_mapping)
```
